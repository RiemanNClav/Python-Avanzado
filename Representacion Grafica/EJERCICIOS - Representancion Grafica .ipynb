```python
class Numeros_Complejos():
    def __init__(self, real, imaginaria):
        self.r = real
        self.i = imaginaria
        
    def mostrar(self):
        if self.i > 0:
            return "El número complejo es: {}+{}i".format(self.r,self.i)
        elif self.i == 0:
            return "El número complejo es: {}".format(self.r)
        else:
            return "El número complejo es: {}{}i".format(self.r,self.i)
        
    @staticmethod
    def suma(complejo1,complejo2):
        real = complejo1.r + complejo2.r
        imag = complejo1.i + complejo2.i
        return (real,imag)
    
    @staticmethod
    def resta(complejo1,complejo2):
        real = complejo1.r - complejo2.r
        imag = complejo1.i - complejo2.i
        return (real,imag)
     
    @staticmethod    
    def producto(complejo1,complejo2):
        real = (complejo1.r*complejo2.r)-(complejo1.i*complejo2.i)
        imag = (complejo1.r*complejo2.i)+(complejo1.i*complejo2.r)
        return (real,imag)
    
    @staticmethod
    def division(complejo1,complejo2):
        real = ((complejo1.r*complejo2.r)+(complejo1.i*complejo2.i))/(complejo2.r**2 + complejo2.i**2)
        imag = ((complejo1.r*complejo2.i)+(complejo1.i*complejo2.r))/(complejo2.r**2 + complejo2.i**2)
        return (real,imag)
    
    @classmethod
    def complejo(clase,real,imag):
        return clase(real,imag)
        

        
        
```

```python
lista_Complejos = []

def IntroducirComplejo():
    real = int(input("Real: "))
    imag = int(input("Imaginaria: "))
    objComplejo = Numeros_Complejos(real,imag)
    lista_Complejos.append(objComplejo)
    
def posicion(z1,signo,a):
    i = 0
    if signo == "+" and a == 2:
        for c in z1:
            if "+" == c:
                return i 
                break
            i += 1    
    elif signo == "-" and a == 2:
        for c in z1:
            if "-" == c and i >= 1:
                return i 
                break 
            i += 1
            
def encontrar(a):
    if a.count("+") == 1:
        return "+"
    elif a.count("-") == 1:
        return "-"
    elif a.count("-") == 2:
        return "-"
        
    
def MostrarComplejo():
    complejo = input("Introduce el número complejo: ")
    z1 = complejo
    signo = encontrar(z1)
    b = z1.find("i")
    a = posicion(z1,signo,2)
    real = int(z1[0:a])
    imaginaria = int(z1[a:b])
    
    for objComplejo in lista_Complejos:
        if real == objComplejo.r and imaginaria == objComplejo.i:
            print(objComplejo.mostrar())
            
def ConsultarComplejo():
    i=0
    for objComplejo in lista_Complejos:
        i+=1
        if objComplejo.i >= 0:
            print("{}.{}+{}i".format(i,objComplejo.r,objComplejo.i))
        else:
            print("{}.{}{}i".format(i,objComplejo.r,objComplejo.i))
            

def IdentificarComplejo(complejo):
    z1 = complejo
    signo = encontrar(z1)
    b = z1.find("i")
    a = posicion(z1,signo,2)
    real = int(z1[0:a])
    imaginaria = int(z1[a:b])
        
    for objComplejo in lista_Complejos:
        if real == objComplejo.r and imaginaria == objComplejo.i:
            return objComplejo
            
def Opciones(opcion,a,b):
    if opcion == 1:
        real = Numeros_Complejos.suma(a,b)[0]
        imag = Numeros_Complejos.suma(a,b)[1]
        Num = Numeros_Complejos.complejo(real,imag)
        return Num
    elif opcion == 2:
        real = Numeros_Complejos.resta(a,b)[0]
        imag = Numeros_Complejos.resta(a,b)[1]
        Num = Numeros_Complejos.complejo(real,imag)
        return Num
    elif opcion == 3:
        real = Numeros_Complejos.producto(a,b)[0]
        imag = Numeros_Complejos.producto(a,b)[1]
        Num = Numeros_Complejos.complejo(real,imag)
        return Num
    elif opcion == 4:
        real = Numeros_Complejos.division(a,b)[0]
        imag = Numeros_Complejos.division(a,b)[1]
        Num = Numeros_Complejos.complejo(real,imag)
        return Num
    
def Operaciones():
    
    print("Escoge 2 números complejos:  ")
    ConsultarComplejo()
    print("")
    a = input(": ")
    b = input(": ")
    Num2 = IdentificarComplejo(a)
    Num3 = IdentificarComplejo(b)
    print("")
    print("Escoge la Operación: ")
    print("1.- Suma")
    print("2.- Resta")
    print("3.- Producto")
    print("4.- Division")
    print("")
    opcion = int(input(": "))
    Num4 = Opciones(opcion,Num2,Num3)
    print(Num4.mostrar())
    lista_Complejos.append(Num4)
            
    
def GraficaComplejo():
    import matplotlib.pyplot as plt 
    import numpy as np 
    lista_complejos = []
    real = []
    imaginaria = []
    ConsultarComplejo()
    n = int(input("Escoge los cantidad de números complejos que quieres graficar: "))
    print("Escribe los números complejos")
    print("")
    for i in range(n):
        a = input(": ")
        Num2 = IdentificarComplejo(a)
        lista_complejos.append(Num2)
    
    for z in lista_complejos:
        real.append(z.r)
        imaginaria.append(z.i)
        
    plt.figure(figsize = (10,10))
    y = np.linspace(-100,100,100)
    y0 = np.linspace(0,0,100)
    x = y
    x0 = y0
    plt.plot(y,y0, c = 'black', alpha = 0.5)
    plt.plot(x0,x, c = 'black', alpha = 0.5)
    plt.scatter(real, imaginaria, c = 'red', s = 120)
    plt.xlim(-10,10)
    plt.ylim(-10,10)
    plt.title("Plano Complejo")
    plt.xlabel("Parte Real")
    plt.ylabel("Parte Imaginaria")
    plt.plot()
    
def main():
        print("\n")
        print("|*************************|")
        print("|*************************|")
        print("")
        while True:
            print("")
            print("Selecciona una de las siguientes opciones: ")
            print("1.- Introduce un Número Complejo")
            print("2.- Mostrar el Número Complejo")
            print("3.- Consultar todos los Números Complejos registrados")
            print("4.- Graficar el/los Número/s Complejo/s")
            print("5.- Hacer operación con 2 Números Complejos")
            print("6.- Salir\n")
            
            opcion = int(input("Opcion: "))
            if opcion == 1:
                IntroducirComplejo()
            elif opcion == 2:
                MostrarComplejo()
            elif opcion == 3:
                ConsultarComplejo()
            elif opcion == 4:
                GraficaComplejo()
            elif opcion == 5:
                Operaciones()
            elif opcion == 6:
                break

    
        
    
            
    
```

Ejercicio 4


Dibuja la función $\frac{sin(x)}{x}$ de color #f06741, la función $\frac{2sin(x)}{x}$ de color #a4f041 y la función $\frac{-sin(x)}{x}$ de color #8d41f0 donde x ∈ $[−5π, 5π]$ en el mismo plot. Crea el array x de tamaño 100 con el método .linspace() de numpy. Haz que el tamaño de la figure sea 8 x 6.


```python
def seno(x):
    import math 
    y = (math.sin(x))
    return y 
    
```

```python
x1 = np.linspace(-5*math.pi,5*math.pi,100)
y1 = list(map(lambda y: seno(y)/y, x1))
y2 = list(map(lambda y: (2*seno(y))/y, x1))
y3 = list(map(lambda y: (-seno(y))/y, x1))
```

```python
plt.figure(figsize = (8,6))
plt.plot(x1, y1, c = '#f06741')
plt.plot(x1,y2, c = '#a4f041')
plt.plot(x1,y3, c = '#8d41f0')
plt.title("Funcion seno(x) / x")
plt.xlabel("[-5pi,5pi]")
plt.ylabel("f(x)")
```
